### 1.2. Trò chơi tổng bằng không (Zero-sum Game)
#### 1.2.1. Định nghĩa và bản chất
<div style="text-align: justify;">
&emsp;&emsp;Thuật ngữ "Trò chơi tổng bằng không" (Zero-sum Game) mô tả một hệ thống trong đó lợi ích của các bên tham gia hoàn toàn đối lập nhau. Theo mô hình này, bất kỳ lợi ích nào mà một người chơi đạt được đều tương đương với thiệt hại của người chơi còn lại. Nói một cách khác, khi cộng tổng giá trị lợi ích của tất cả người chơi ở bất kỳ trạng thái kết thúc nào, kết quả luôn bằng không (hoặc một hằng số không đổi). Biểu diễn bằng công thức toán học: nếu $U_A$ là hàm lợi ích của người chơi A và $U_B$ là hàm lợi ích của người chơi B, thì trong trò chơi tổng bằng không hai người, ta có: <br><br>
$$U_A(s) + U_B(s) = 0, \quad \forall s \in S$$
<br>
&emsp;&emsp;(Với $s$ là một trạng thái bất kỳ trong không gian trạng thái $S$). <br>
&emsp;&emsp;Từ đây suy ra rằng trong loại trò chơi này không tồn tại khả năng hợp tác. Bất cứ khi nào người chơi A (MAX) cố gắng tăng điểm số của mình, điều đó đồng nghĩa với việc đẩy người chơi B (MIN) vào tình thế bất lợi. Nguyên lý này là nền tảng để phát triển thuật toán Minimax: một phía tìm cách đạt giá trị lớn nhất, phía kia tìm cách giảm giá trị xuống mức thấp nhất.
</div>

#### 1.2.2. Các đặc trưng cơ bản của môi trường trò chơi
<div style="text-align: justify;">
&emsp;&emsp;Để các thuật toán tìm kiếm đối kháng như Minimax và Alpha-Beta có thể hoạt động hiệu quả, trò chơi cần đáp ứng những điều kiện kỹ thuật nhất định. Báo cáo này xem xét các bài toán có ba đặc điểm sau:<br>
&emsp;&emsp;&emsp;1. Thông tin hoàn hảo (Perfect Information): Ở mọi thời điểm, cả hai người chơi đều biết rõ toàn bộ tình trạng của môi trường (bao gồm vị trí các quân cờ trên bàn). Không có thông tin bị che giấu như lá bài của đối thủ trong Poker hay tầm nhìn bị hạn chế trong các game chiến thuật thời gian thực. Đặc điểm này giúp thuật toán có thể xây dựng cây tìm kiếm một cách chính xác mà không cần đến các kỹ thuật xác suất để dự đoán thông tin ẩn.<br>
&emsp;&emsp;&emsp;2. Môi trường tất định (Deterministic): Mỗi hành động luôn dẫn đến một kết quả duy nhất, không có yếu tố ngẫu nhiên (chẳng hạn như việc tung xúc xắc). Khi người chơi thực hiện nước đi $m$ từ trạng thái A, trạng thái tiếp theo B được xác định hoàn toàn. Điều này đảm bảo rằng chiến lược do AI đề xuất sẽ không bị ảnh hưởng bởi các yếu tố may rủi. <br>
&emsp;&emsp;&emsp;3. Lượt chơi luân phiên (Turn-based): Trò chơi tiến hành theo các bước rời rạc, hai người chơi lần lượt thực hiện hành động của mình. Giữa các lượt, trạng thái trò chơi không thay đổi cho đến khi một người chơi thực hiện nước đi.
</div>
<div style="text-align: justify;">

> **Ghi chú:** Đối với những trò chơi phức tạp như Cờ Caro, việc tính toán đến tận cùng để xác định giá trị +1 hoặc -1 là không khả thi về mặt tính toán. Vì vậy, khái niệm "tổng bằng không" được mở rộng sang các giá trị heuristic tại các trạng thái trung gian, trong đó giá trị dương biểu thị lợi thế cho MAX và giá trị âm biểu thị lợi thế cho MIN.
</div>


### 1.3. Các thành phần của mô hình trò chơi
<div>
&emsp;&emsp;Việc biến một trò chơi thực tế thành một bài toán có thể giải quyết bằng máy tính đòi hỏi phải mô hình hóa nó thông qua bốn yếu tố cơ bản, các yếu tố này kết hợp với nhau để tạo nên không gian trạng thái:<br>
&emsp;&emsp;&emsp;1. Trạng thái khởi đầu ($S_0$): Đây là cấu hình ban đầu của bàn cờ (có thể là bàn cờ trống hoặc có các quân cờ ở vị trí mặc định) kèm theo thông tin về người chơi nào được đi trước. Trạng thái này đóng vai trò là nút gốc trong cây trò chơi.<br>
&emsp;&emsp;&emsp;2. Hàm sinh trạng thái kế tiếp (Successor Function): Chức năng này quy định những nước đi nào được phép thực hiện. Đối với mỗi trạng thái $s$, hàm sẽ trả về một tập các cặp $(a, s')$, trong đó $a$ là hành động và $s'$ là trạng thái mới sau khi thực hiện hành động đó.<br>
&emsp;&emsp;&emsp;3. Hàm kiểm tra kết thúc (Terminal Test): Đây là một hàm logic (Boolean) để xác định xem trò chơi đã kết thúc hay chưa, có thể là do một bên thắng, thua, hoặc hòa.<br>
&emsp;&emsp;&emsp;4. Hàm lợi ích (Utility Function): Hàm này gán cho mỗi trạng thái kết thúc một con số cụ thể (chẳng hạn: +1 nếu thắng, -1 nếu thua, 0 nếu hòa) nhằm đánh giá và so sánh các kết quả khác nhau.
</div>


### 1.4. Phân tích không gian trạng thái
<div>
&emsp;&emsp;Số lượng trạng thái trong các trò chơi đối kháng tăng lên theo cấp số mũ với công thức $O(b^d)$, trong đó $b$ đại diện cho hệ số nhánh (số nước đi trung bình tại mỗi trạng thái) và $d$ là độ sâu của ván cờ. Đối với những trò chơi phức tạp như Cờ vua hoặc Cờ tướng, tổng số trạng thái có thể đạt đến con số khổng lồ $10^{123}$. Với quy mô không gian như vậy, phương pháp duyệt toàn bộ (Brute-force) để tìm ra kết quả chính xác là không thể thực hiện được do giới hạn về thời gian và khả năng xử lý của máy tính. Chính vì lý do này, những phương pháp tối ưu như cắt tỉa Alpha-Beta và các hàm đánh giá heuristic đóng vai trò cực kỳ quan trọng.
</div>


## 2. Thuật toán Minimax


### 2.1. Nguyên lý hoạt động của Minimax

#### 2.1.1. Cơ chế
Minimax là một giải thuật đệ quy nhằm tìm kiếm quyết định tốt nhất trong các trò chơi đối kháng (zero-sum games) với điều kiện thông tin hoàn hảo. Ý tưởng chính của thuật toán xuất phát từ giả thiết rằng **đối thủ cũng sẽ chơi một cách tối ưu** (tức là đối thủ luôn lựa chọn nước đi có hại nhất đối với ta).

Hai người chơi trong mô hình này được gọi là:
* **MAX (Máy tính/AI):** Người chơi này luôn tìm cách **tăng tối đa** (maximize) giá trị lợi ích của mình.
* **MIN (Đối thủ):** Người chơi này luôn tìm cách **giảm tối thiểu** (minimize) giá trị lợi ích của MAX (nghĩa là khiến MAX thua hoặc đạt điểm thấp nhất có thể).

#### 2.1.2. Chiến lược "Nhìn trước"
Minimax không chỉ đánh giá nước đi ở thời điểm hiện tại mà còn mô phỏng các tình huống có thể xảy ra trong tương lai bằng cách duyệt sâu vào cây trò chơi (Game Tree).
* Giải thuật sẽ khám phá cây trò chơi đến một mức độ sâu nhất định hoặc cho đến khi gặp trạng thái kết thúc (Terminal state).
* Ở các nút lá, hàm đánh giá (Evaluation function) hoặc hàm lợi ích (Utility function) sẽ tính toán và gán điểm số cho trạng thái đó.
* Điểm số này sau đó được truyền ngược lên (back-propagation) từ các nút lá về nút gốc, giúp MAX có cơ sở để đưa ra quyết định ở trạng thái hiện tại.


### 2.2. Biểu diễn toán học của Minimax

Để có thể mô tả thuật toán một cách chính xác, chúng ta sử dụng các ký hiệu toán học để đại diện cho các thành phần trong cây trò chơi.

Định nghĩa các ký hiệu:
* $s$: Trạng thái hiện tại trên bàn cờ.
* $PLAYER(s)$: Hàm cho biết người chơi nào đang có lượt tại trạng thái $s$ (MAX hay MIN).
* $ACTIONS(s)$: Tập hợp tất cả các nước đi hợp lệ có thể thực hiện từ trạng thái $s$.
* $RESULT(s, a)$: Trạng thái mới xuất hiện sau khi thực hiện hành động $a$ tại trạng thái $s$.
* $UTILITY(s)$: Hàm lợi ích trả về một giá trị số tại các trạng thái kết thúc.

Giá trị Minimax của trạng thái $s$, ký hiệu $V(s)$ hoặc $Minimax(s)$, được định nghĩa theo cách đệ quy như sau:

$$
V(s) =
\begin{cases}
UTILITY(s) & \text{nếu } TERMINAL\_TEST(s) \text{ là true} \\
\max_{a \in ACTIONS(s)} V(RESULT(s, a)) & \text{nếu } PLAYER(s) = MAX \\
\min_{a \in ACTIONS(s)} V(RESULT(s, a)) & \text{nếu } PLAYER(s) = MIN
\end{cases}
$$

**Ý nghĩa của công thức:**
1.  **Trường hợp cơ sở (Base case):** Khi $s$ là trạng thái kết thúc, giá trị của nó được xác định trực tiếp bởi kết quả của trò chơi (ví dụ: +1 nếu thắng, -1 nếu thua, 0 nếu hòa).
2.  **Lượt của MAX:** MAX sẽ xem xét tất cả các trạng thái con $s'$ (sinh ra từ các hành động $a$ khả thi) và lựa chọn giá trị **cao nhất** trong số chúng.
3.  **Lượt của MIN:** MIN sẽ xem xét tất cả các trạng thái con $s'$ và lựa chọn giá trị **thấp nhất** trong số chúng (nhằm hạn chế lợi ích của MAX).


### 2.3. Quy trình thực hiện

Thuật toán Minimax được thực hiện theo phương pháp tìm kiếm theo chiều sâu (Depth-First Search - DFS) với năm giai đoạn chính:

**Bước 1: Khởi tạo và Tạo cây**
* Xuất phát từ trạng thái hiện tại (được coi là nút gốc).
* Tạo ra các trạng thái con hợp lệ từ nút gốc. Quá trình này tiếp tục cho đến khi đạt đến giới hạn độ sâu hoặc gặp trạng thái kết thúc trò chơi.

**Bước 2: Đánh giá tại nút lá**
* Khi thuật toán đi đến các nút cuối cùng của cây tìm kiếm (nút lá), nó sử dụng hàm đánh giá (Heuristic hoặc Utility) để tính toán điểm số cho trạng thái đó.
* Trong trò chơi Tic-Tac-Toe: Nếu MAX thắng thì gán +1, nếu MIN thắng thì gán -1, nếu hòa thì gán 0.

**Bước 3: Lan truyền ngược**
* Các giá trị từ nút lá được truyền ngược lên nút cha ngay phía trên.

**Bước 4: Chọn lọc giá trị**
* Ở mỗi nút cha, thuật toán sẽ so sánh các giá trị nhận được từ các nút con:
    * Nếu nút cha thuộc về **MAX**: Nó sẽ chọn giá trị lớn nhất ($v = \max(v, \text{child\_value})$).
    * Nếu nút cha thuộc về **MIN**: Nó sẽ chọn giá trị nhỏ nhất ($v = \min(v, \text{child\_value})$).

**Bước 5: Ra quyết định**
* Việc truyền giá trị ngược lên tiếp tục cho đến khi quay trở lại nút gốc.
* Tại nút gốc (lượt của MAX), AI sẽ quyết định chọn **hành động $a$** dẫn đến nút con có giá trị Minimax lớn nhất.

**Ví dụ minh họa quy trình:**
Xét một cây trò chơi đơn giản:
1.  Các nút lá D, E, F có giá trị tương ứng là 2, 3, 4 (giả sử đây là lượt MAX đánh giá).
2.  Nút cha của chúng là B (lượt MIN). MIN sẽ chọn $\min(2, 3, 4) = 2$. Do đó giá trị của nút B là 2.
3.  Tương tự, giả sử nhánh còn lại có nút C với giá trị là 5.
4.  Tại nút gốc A (lượt MAX), MAX so sánh B (2) và C (5). MAX chọn $\max(2, 5) = 5$.
5.  Kết luận: AI quyết định đi theo nhánh dẫn đến C.


### 2.4. Phát biểu chi tiết bài toán
Nhiệm vụ của bài toán là thiết kế và xây dựng một tác tử thông minh (AI Agent) có thể tham gia vào trò chơi **Tic-Tac-Toe** (hoặc mở rộng sang các trò chơi như Cờ Caro/Gomoku) với tư cách là một người chơi có thể đấu với con người hoặc với một AI khác.

  * **Bối cảnh:** Trò chơi được tổ chức trên một bàn cờ có kích thước $N \times N$ (ví dụ: Tic-Tac-Toe sử dụng bàn cờ $3 \times 3$, còn Caro thường dùng bàn cờ $10 \times 10$ hoặc lớn hơn).
  * **Mục tiêu của AI:** Tìm ra một chiến lược đặt quân sao cho đạt được kết quả tối ưu nhất: **Thắng** (trong trường hợp đối thủ mắc sai lầm) hoặc **Hòa** (khi đối thủ chơi một cách tối ưu), và đảm bảo không bao giờ để thua.


### 2.5. Các thành phần của mô hình bài toán

Theo lý thuyết tìm kiếm đối kháng, bài toán được định nghĩa hình thức thông qua 5 thành phần sau:

1.  **Trạng thái ban đầu ($S_0$):** Bàn cờ rỗng kích thước $3 \times 3$, quy định người chơi X được đi trước.
2.  **Người chơi ($PLAYER(s)$):** Hàm xác định tại trạng thái $s$ hiện tại, lượt đi thuộc về ai (MAX hay MIN).
3.  **Hành động ($ACTIONS(s)$):** Tập hợp các nước đi hợp lệ $(i, j)$ mà người chơi có thể thực hiện tại trạng thái $s$ (là các ô còn trống).
4.  **Mô hình chuyển trạng thái ($RESULT(s, a)$):** Hàm trả về trạng thái bàn cờ mới sau khi thực hiện hành động $a$ lên trạng thái $s$.
5.  **Hàm kiểm tra kết thúc ($TERMINAL(s)$):** Hàm Boolean trả về `True` nếu trò chơi kết thúc (có hàng/cột/chéo đủ quân hoặc hết ô trống), ngược lại `False`.


### 2.6. Thiết kế và lựa chọn cấu trúc dữ liệu

#### 2.6.1. Biểu diễn bàn cờ (State Representation)

Việc lựa chọn cấu trúc dữ liệu ảnh hưởng trực tiếp đến tốc độ sinh nước đi và kiểm tra kết quả.

  * **Lựa chọn:** Sử dụng **Mảng 2 chiều (2D Array/Matrix)** hoặc **Danh sách lồng nhau (List of Lists)**.
      * Ví dụ: `board[3][3]`
  * **Lý giải:**
      * *Trực quan:* Ánh xạ 1-1 với bàn cờ thực tế, dễ dàng truy xuất qua tọa độ dòng `i` và cột `j`.
      * *Hiệu quả kiểm tra:* Dễ dàng duyệt vòng lặp để kiểm tra thắng thua theo hàng ngang, dọc và đường chéo.
      * *Tính bất biến:* Khi truyền vào đệ quy, cần sử dụng cơ chế sao chép sâu (`deepcopy`) để đảm bảo trạng thái cha không bị ảnh hưởng bởi phép thử ở trạng thái con.

#### 2.6.2. Biểu diễn giá trị ô cờ

  * **Lựa chọn:** Sử dụng chuỗi ký tự `"X"`, `"O"` và `None` (hoặc hằng số `EMPTY`).
  * **Lý giải:** Phù hợp với tư duy lập trình hướng đối tượng và dễ hiển thị ra giao diện (Console/UI). Trong các cài đặt tối ưu hơn (như C++), có thể dùng số nguyên (1, -1, 0) để tính tổng hàng/cột nhanh hơn, nhưng với Python, chuỗi ký tự là lựa chọn tốt nhất.


### 2.7. Đầu vào và Đầu ra của thuật toán

  * **Đầu vào (Input):**
      * `state`: Trạng thái bàn cờ hiện tại (một ma trận chứa các quân X, O, Empty).
      * `depth` (Tùy chọn): Độ sâu giới hạn nếu áp dụng cắt tỉa hoặc bàn cờ lớn.
      * `alpha`, `beta` (Nếu dùng Alpha-Beta): Các giá trị giới hạn để cắt tỉa nhánh.
  * **Đầu ra (Output):**
      * `best_move`: Tọa độ `(row, col)` của nước đi tối ưu nhất.
      * `value`: Giá trị điểm số dự kiến (ví dụ: +1 nếu chắc thắng, 0 nếu hòa).


### 2.8. Thiết kế Hàm đánh giá (Utility & Heuristic Function)

#### Trường hợp duyệt đến cùng (Tic-Tac-Toe 3x3)

Do không gian trạng thái nhỏ, ta có thể duyệt đến tận cùng (Terminal Node). Khi đó hàm đánh giá chính là **Hàm lợi ích (Utility Function)**:

  * Nếu $s$ là trạng thái MAX thắng: Trả về **+1** (hoặc +10).
  * Nếu $s$ là trạng thái MIN thắng: Trả về **-1** (hoặc -10).
  * Nếu $s$ là Hòa: Trả về **0**.

### 2.9. Mã giả thuật toán (Pseudocode)

#### Hàm quyết định nước đi (Minimax-Decision)

```plaintext
function MINIMAX-DECISION(state) returns an action
    v ← MAX-VALUE(state)
    return action trong SUCCESSORS(state) có giá trị v
```
#### Hàm tìm giá trị cực đại (Max-Value) - Dành cho lượt AI

```plaintext
function MAX-VALUE(state) returns a utility value
    if TERMINAL-TEST(state) then
        return UTILITY(state)
    
    v ← -∞
    for each (action, state') in SUCCESSORS(state) do
        v ← MAX(v, MIN-VALUE(state'))
    return v
```
#### Hàm tìm giá trị cực tiểu (Min-Value) - Dành cho lượt Đối thủ

```plaintext
function MIN-VALUE(state) returns a utility value
    if TERMINAL-TEST(state) then
        return UTILITY(state)
    
    v ← +∞
    for each (action, state') in SUCCESSORS(state) do
        v ← MIN(v, MAX-VALUE(state'))
    return v
```

**Triển khai thuật toán minimax vào trò chơi tictactoe 3x3 <br>
Bước 1: Khai báo thư viện và các hằng số cơ bản cho trò chơi.**


````python
import copy
import math
import numpy as np

X = "X"
O = "O"
EMPTY = None

N = 3
WIN_COUNT = 3

def initial_state():
    return [[EMPTY for _ in range(N)] for _ in range(N)]
````

**Bước 2: Định nghĩa các hàm xử lý luật chơi cơ bản**

Hàm `player(board)` xác định lượt đi hiện tại dựa trên số lượng quân cờ trên bàn. Nguyên tắc hoạt động rất đơn giản: đếm số lượng quân X và O, nếu số lượng bằng nhau thì đến lượt X (vì X luôn đi trước), còn nếu X nhiều hơn O thì đến lượt O. Hàm này đảm bảo tính công bằng trong việc phân bổ lượt chơi.

Hàm `actions(board)` có nhiệm vụ quét toàn bộ bàn cờ để tìm ra tất cả các vị trí còn trống. Kết quả trả về là một tập hợp (set) chứa các tuple (hàng, cột) đại diện cho các ô có thể đánh. Việc sử dụng set giúp loại bỏ các phần tử trùng lặp và tăng tốc độ kiểm tra sau này.

Hàm `result(board, action)` là một hàm quan trọng tạo ra trạng thái mới sau khi thực hiện nước đi. Điểm đặc biệt là hàm này sử dụng `copy.deepcopy()` để tạo một bản sao hoàn toàn độc lập của bàn cờ, đảm bảo rằng bàn cờ gốc không bị thay đổi trong quá trình đệ quy của thuật toán Minimax.


````python
def player(board):
    count_x = 0
    count_o = 0
    for row in board:
        for cell in row:
            if cell == X:
                count_x += 1
            elif cell == O:
                count_o += 1

    if count_x > count_o:
        return O
    else:
        return X

def actions(board):
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    (i, j) = action
    if board[i][j] is not EMPTY:
        raise Exception("Nước đi không hợp lệ: Ô đã có quân.")

    curr_player = player(board)
    result_board = copy.deepcopy(board)
    result_board[i][j] = curr_player
    return result_board
````

**Bước 3: Định nghĩa các hàm kiểm tra kết thúc và đánh giá**

Hàm `check_line(board, player)` là một hàm phụ trợ quan trọng kiểm tra xem một người chơi có đạt đủ số quân liên tiếp (WIN_COUNT) để thắng hay không. Hàm này quét bàn cờ theo bốn hướng: hàng ngang, hàng dọc, đường chéo chính và đường chéo phụ. Với mỗi hướng, hàm sử dụng vòng lặp lồng nhau để kiểm tra từng đoạn có độ dài WIN_COUNT, sử dụng hàm `all()` để đảm bảo tất cả các ô trong đoạn đều thuộc về cùng một người chơi.

Hàm `winner(board)` gọi `check_line` cho cả hai người chơi X và O để xác định người thắng cuộc. Hàm `terminal(board)` xác định xem trò chơi đã kết thúc hay chưa. Trò chơi được coi là kết thúc trong hai trường hợp: một là đã có người thắng (kiểm tra qua hàm winner), hai là bàn cờ đã đầy (không còn ô trống nào). Hàm này đóng vai trò quan trọng trong việc quyết định điểm dừng của thuật toán đệ quy.

Hàm `utility(board)` là hàm lợi ích gán điểm số cho các trạng thái kết thúc. Giá trị +1 được gán khi X thắng (MAX thắng), -1 khi O thắng (MIN thắng), và 0 khi hòa. Đây là giá trị cơ sở mà thuật toán Minimax sử dụng để đánh giá và so sánh các nhánh trong cây trò chơi.


````python
def check_line(board, player):
    for r in range(N):
        for c in range(N - WIN_COUNT + 1):
            if all(board[r][c+k] == player for k in range(WIN_COUNT)): return True

    for c in range(N):
        for r in range(N - WIN_COUNT + 1):
            if all(board[r+k][c] == player for k in range(WIN_COUNT)): return True

    for r in range(N - WIN_COUNT + 1):
        for c in range(N - WIN_COUNT + 1):
            if all(board[r+k][c+k] == player for k in range(WIN_COUNT)): return True

    for r in range(N - WIN_COUNT + 1):
        for c in range(WIN_COUNT - 1, N):
            if all(board[r+k][c-k] == player for k in range(WIN_COUNT)): return True

    return False

def winner(board):
    if check_line(board, X): return X
    if check_line(board, O): return O
    return None

def terminal(board):
    if winner(board) is not None:
        return True

    for row in board:
        if EMPTY in row:
            return False
    return True

def utility(board):
    win_player = winner(board)
    if win_player == X:
        return 1
    elif win_player == O:
        return -1
    else:
        return 0
````

**Bước 4: Định nghĩa giải thuật Minimax**

Hàm `maxValue(state)` được thiết kế để tìm giá trị điểm số lớn nhất có thể cho lượt của MAX. Hàm này hoạt động theo nguyên lý đệ quy: đầu tiên kiểm tra xem trạng thái hiện tại đã kết thúc chưa, nếu có thì trả về giá trị utility. Nếu chưa, hàm sẽ duyệt qua tất cả các nước đi có thể, gọi đệ quy hàm `minValue` cho mỗi nước đi (vì sau lượt MAX là lượt MIN), và chọn giá trị lớn nhất trong số đó. Việc khởi tạo `v = -math.inf` đảm bảo rằng bất kỳ giá trị nào cũng sẽ lớn hơn giá trị ban đầu.

Hàm `minValue(state)` hoạt động tương tự nhưng theo hướng ngược lại - tìm giá trị nhỏ nhất cho MIN. Hàm này cũng kiểm tra trạng thái kết thúc trước, sau đó duyệt các nước đi và gọi đệ quy `maxValue`, cuối cùng chọn giá trị nhỏ nhất. Khởi tạo `v = math.inf` để đảm bảo mọi giá trị đều nhỏ hơn giá trị ban đầu.

Hàm `minimax(board)` là hàm chính quyết định nước đi tối ưu. Hàm này xác định người chơi hiện tại, sau đó thử từng nước đi có thể và đánh giá chúng thông qua việc gọi `maxValue` hoặc `minValue` tùy thuộc vào lượt chơi. Nước đi nào cho giá trị tốt nhất sẽ được chọn làm nước đi cuối cùng. Đây chính là điểm kết thúc của quá trình tìm kiếm trong cây trò chơi.


````python
def maxValue(state):
    if terminal(state):
        return utility(state)

    v = -math.inf

    for action in actions(state):
        v = max(v, minValue(result(state, action)))

    return v

def minValue(state):
    if terminal(state):
        return utility(state)

    v = math.inf

    for action in actions(state):
        v = min(v, maxValue(result(state, action)))

    return v

def minimax(board):
    if terminal(board):
        return None

    curr_player = player(board)

    if curr_player == X:
        best_score = -math.inf
        best_move = None

        for action in actions(board):
            score = minValue(result(board, action))

            if score > best_score:
                best_score = score
                best_move = action
        return best_move

    else:
        best_score = math.inf
        best_move = None

        for action in actions(board):
            score = maxValue(result(board, action))

            if score < best_score:
                best_score = score
                best_move = action
        return best_move
````

**Bước 5: Xây dựng chương trình chính**

Chương trình chính được thiết kế để tạo ra một trải nghiệm chơi game tương tác giữa người dùng và máy tính. Đầu tiên, chương trình yêu cầu người dùng nhập kích thước bàn cờ N, sau đó khởi tạo bàn cờ trống. Tiếp theo, người dùng được chọn phe chơi (X hoặc O), với X luôn được quyền đi trước. Vòng lặp chính của trò chơi sẽ tiếp tục cho đến khi có người thắng hoặc hòa. Trong mỗi lượt, nếu là lượt của người chơi, chương trình sẽ yêu cầu nhập tọa độ và kiểm tra tính hợp lệ. Nếu là lượt của máy, thuật toán Minimax sẽ được gọi để tính toán nước đi tối ưu. Khi trò chơi kết thúc, kết quả sẽ được hiển thị rõ ràng cho người chơi.


````python
3
if __name__ == "__main__":
    # 0. Nhập kích thước bàn cờ và số quân cần để thắng
    global N, WIN_COUNT
    while True:
        try:
            N = int(input("Nhập kích thước bàn cờ N (ví dụ: 3 cho 3x3, 5 cho 5x5): "))
            if N >= 3:
                break
            else:
                print("Kích thước bàn cờ phải lớn hơn hoặc bằng 3. Vui lòng nhập lại.")
        except ValueError:
            print("Vui lòng nhập một số nguyên hợp lệ.")

    while True:
        try:
            WIN_COUNT = int(input(f"Nhập số quân liên tiếp cần để thắng (tối đa {N}): "))
            if 3 <= WIN_COUNT <= N:
                break
            else:
                print(f"Số quân cần để thắng phải từ 3 đến {N}. Vui lòng nhập lại.")
        except ValueError:
            print("Vui lòng nhập một số nguyên hợp lệ.")

    # 1. Khởi tạo bàn cờ
    board = initial_state()

    print("------------------------------------------")
    print(f"   GAME TIC-TAC-TOE {N}x{N}: NGƯỜI vs MÁY (AI)")
    print(f"   Luật thắng: {WIN_COUNT} quân liên tiếp")
    print("------------------------------------------")

    # 2. Chọn phe
    while True:
        user_side = input("Bạn muốn chơi X hay O? (Lưu ý: X luôn đi trước): ").upper()
        if user_side in [X, O]:
            break
        print("Vui lòng chỉ nhập X hoặc O.")

    if user_side == X:
        ai_side = O
        print(f"-> Bạn chọn X. Bạn đi trước!")
    else:
        ai_side = X
        print(f"-> Bạn chọn O. Máy (X) đi trước!")

    # 3. Vòng lặp trò chơi
    while True:
        # Kiểm tra xem game kết thúc chưa
        game_over = terminal(board)
        curr_player = player(board)

        # Nếu kết thúc, in kết quả và thoát
        if game_over:
            print("\n=== TRÒ CHƠI KẾT THÚC ===")
            print(np.array(board))
            win = winner(board)
            if win is None:
                print(">> KẾT QUẢ: HÒA (Draw)!")
            elif win == user_side:
                print(">> KẾT QUẢ: BẠN THẮNG (Không thể xảy ra với Minimax chuẩn)!")
            else:
                print(">> KẾT QUẢ: MÁY THẮNG!")
            break

        # --- LƯỢT CỦA NGƯỜI CHƠI ---
        if curr_player == user_side:
            print(f"\n--- Lượt của BẠN ({user_side}) ---")
            print(np.array(board)) # In bàn cờ cho người dùng xem

            while True:
                try:
                    print(f"Nhập tọa độ nước đi (0-{N-1}):")
                    r = int(input("   Hàng: "))
                    c = int(input("   Cột : "))

                    if (r, c) in actions(board):
                        board = result(board, (r, c))
                        break
                    else:
                        print("!! Ô này không hợp lệ hoặc đã có người đánh. Thử lại.")
                except ValueError:
                    print("!! Vui lòng nhập số nguyên.")

        # --- LƯỢT CỦA MÁY (AI) ---
        else:
            print(f"\n--- Lượt của MÁY ({ai_side}) đang suy nghĩ... ---")
            move = minimax(board)
            board = result(board, move)
            print(f"Máy đã đánh vào ô: {move}")
````

## 3. Cắt tỉa Alpha-Beta cho bài toán tictactoe 5x5


### 3.1. Nguyên lý hoạt động
Cắt tỉa Alpha-Beta là một phương pháp tối ưu hóa dành cho thuật toán Minimax. Kỹ thuật này có khả năng loại bỏ một số lượng lớn nút không cần thiết trong cây trò chơi mà **vẫn đảm bảo kết quả cuối cùng giống hệt** như khi sử dụng Minimax thuần túy.

* **Ý tưởng chính:** Khi thuật toán phát hiện một nước đi không tốt (đối với người chơi đang xét) trong một nhánh, nó sẽ dừng việc kiểm tra các nước đi còn lại trong nhánh đó, bởi vì nhánh đó chắc chắn sẽ không được lựa chọn (hoặc đối thủ sẽ không cho phép lựa chọn đó).
* **Hai tham số cắt tỉa:**
    * **Alpha ($\alpha$):** Đây là giá trị tốt nhất (cao nhất) mà người chơi **MAX** đã phát hiện được cho đến lúc này dọc theo đường đi hiện tại. Giá trị $\alpha$ chỉ có thể tăng lên, không bao giờ giảm.
    * **Beta ($\beta$):** Đây là giá trị tốt nhất (thấp nhất) mà người chơi **MIN** đã phát hiện được cho đến lúc này dọc theo đường đi hiện tại. Giá trị $\beta$ chỉ có thể giảm xuống, không bao giờ tăng.

### 3.2. Biểu diễn toán học và điều kiện cắt tỉa

#### Khởi tạo
Ở nút gốc, chúng ta khởi tạo cửa sổ tìm kiếm với phạm vi rộng nhất có thể:
* $\alpha = -\infty$
* $\beta = +\infty$

#### Quy tắc cập nhật
* **Tại nút MAX:** $\alpha = \max(\alpha, \text{giá trị của nút con})$.
* **Tại nút MIN:** $\beta = \min(\beta, \text{giá trị của nút con})$.

#### Điều kiện cắt tỉa
Quá trình cắt tỉa được kích hoạt ngay khi thỏa mãn điều kiện:
$$\alpha \ge \beta$$

**Giải thích:**
* Nếu tại một nút MAX, chúng ta tìm được một giá trị $\ge \beta$ (giá trị tốt nhất mà MIN đã có ở một nhánh khác), thì MIN sẽ không bao giờ cho phép MAX đi vào nhánh này. Vì vậy, MAX không cần phải xem xét thêm các nút con còn lại.
* Tương tự, nếu tại một nút MIN, chúng ta tìm được giá trị $\le \alpha$, MAX sẽ không bao giờ chọn đi vào nhánh này vì MAX đã có một lựa chọn tốt hơn (giá trị $\alpha$) ở một nơi khác.

### 3.3. Quy trình thực hiện chi tiết
Quy trình được thực hiện dựa trên cơ chế duyệt theo chiều sâu (DFS) kết hợp với việc cập nhật các tham số:

1.  **Bước 1:** Bắt đầu từ nút gốc với khoảng giá trị $(\alpha=-\infty, \beta=+\infty)$.
2.  **Bước 2:** Đi xuống nút con đầu tiên bên trái, đồng thời truyền các giá trị $\alpha, \beta$ xuống.
3.  **Bước 3 (Tại nút lá):** Tính toán giá trị hàm lợi ích và truyền ngược lên nút cha.
4.  **Bước 4 (Tại nút cha):**
    * Cập nhật giá trị $\alpha$ (nếu là nút MAX) hoặc $\beta$ (nếu là nút MIN) dựa trên giá trị vừa nhận được từ nút con.
    * Kiểm tra xem điều kiện $\alpha \ge \beta$ có được thỏa mãn hay không.
    * **Nếu thỏa mãn:** Dừng việc duyệt các nút con còn lại của nút này (thực hiện cắt tỉa - Prune).
    * **Nếu không thỏa mãn:** Tiếp tục duyệt nút con tiếp theo với các giá trị $\alpha, \beta$ đã được cập nhật.
5.  **Bước 5:** Lặp lại toàn bộ quá trình cho đến khi quay trở lại nút gốc.

<br>

**Ví dụ minh họa:**
Giả sử tại nút MIN (C), chúng ta đã tìm được một nhánh con có giá trị là 3. Lúc này $\beta$ tại C được cập nhật thành 3 (hoặc một giá trị nhỏ hơn). Trước đó, tại nút cha của C (là nút MAX, gọi là S), chúng ta đã có $\alpha = 10$.
So sánh tại C: $\alpha (10) \ge \beta (3)$.
$\rightarrow$ Thực hiện cắt tỉa! Không cần xem xét các nút con còn lại của C vì S (MAX) chắc chắn sẽ không chọn đường đi qua C (do giá trị $\le 3$ thấp hơn 10).


**Bước 1: Khai báo các thư viện cần thiết và định nghĩa các hằng số cơ bản.**


````python
import copy
import math
import numpy as np

X = "X"
O = "O"
EMPTY = None

N = 5
WIN_COUNT = 4

def initial_state():
    return [[EMPTY for _ in range(N)] for _ in range(N)]
````

**Bước 2: Khai báo các hàm tìm nước đi và xác định lượt chơi**

Các hàm trong phần này hoạt động tương tự như phần Minimax nhưng được tối ưu hóa cho bàn cờ lớn hơn. Hàm `player(board)` vẫn dựa trên nguyên tắc đếm quân để xác định lượt đi. Hàm `actions(board)` sẽ quét qua tất cả các ô trên bàn cờ NxN để tìm các vị trí trống, với bàn cờ 5x5 sẽ có tối đa 25 ô cần kiểm tra. Hàm `result(board, action)` tiếp tục sử dụng deepcopy để đảm bảo tính toàn vẹn dữ liệu trong quá trình đệ quy của thuật toán Alpha-Beta.


````python
def player(board):
    count_x = 0
    count_o = 0
    for row in board:
        for cell in row:
            if cell == X:
                count_x += 1
            elif cell == O:
                count_o += 1

    if count_x > count_o:
        return O
    else:
        return X

def actions(board):
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action

    if result_board[i][j] is not EMPTY:
        raise Exception("Nước đi không hợp lệ: Ô đã có quân.")

    result_board[i][j] = curr_player
    return result_board
````

**Bước 3: Định nghĩa các hàm kiểm tra thắng thua và hàm đánh giá**

Hàm `check_line(board, player)` là một hàm phụ trợ quan trọng kiểm tra xem một người chơi có đạt đủ số quân liên tiếp (WIN_COUNT) để thắng hay không. Hàm này quét bàn cờ theo bốn hướng: hàng ngang, hàng dọc, đường chéo chính và đường chéo phụ. Với mỗi hướng, hàm sử dụng vòng lặp lồng nhau để kiểm tra từng đoạn có độ dài WIN_COUNT, sử dụng hàm `all()` để đảm bảo tất cả các ô trong đoạn đều thuộc về cùng một người chơi.

Hàm `winner(board)` gọi `check_line` cho cả hai người chơi X và O để xác định người thắng cuộc. Hàm `terminal(board)` kiểm tra điều kiện kết thúc bằng cách xem có người thắng hoặc bàn cờ đã đầy chưa.

Hàm `utility(board)` có hai phần chính: phần đầu xử lý trường hợp game đã kết thúc, trả về giá trị tuyệt đối (10000 hoặc -10000) để phân biệt rõ ràng với các giá trị heuristic. Phần thứ hai là hàm đánh giá heuristic cho các trạng thái chưa kết thúc, ở đây sử dụng một heuristic đơn giản là đếm số lượng quân cờ của mỗi bên và trả về hiệu số. Trong thực tế, có thể phát triển heuristic phức tạp hơn như đếm số chuỗi quân liên tiếp, đánh giá vị trí chiến lược, v.v.


````python
def check_line(board, player):
    for r in range(N):
        for c in range(N - WIN_COUNT + 1):
            if all(board[r][c+k] == player for k in range(WIN_COUNT)): return True

    for c in range(N):
        for r in range(N - WIN_COUNT + 1):
            if all(board[r+k][c] == player for k in range(WIN_COUNT)): return True

    for r in range(N - WIN_COUNT + 1):
        for c in range(N - WIN_COUNT + 1):
            if all(board[r+k][c+k] == player for k in range(WIN_COUNT)): return True

    for r in range(N - WIN_COUNT + 1):
        for c in range(WIN_COUNT - 1, N):
            if all(board[r+k][c-k] == player for k in range(WIN_COUNT)): return True

    return False

def winner(board):
    if check_line(board, X): return X
    if check_line(board, O): return O
    return None

def terminal(board):
    if winner(board) is not None:
        return True

    for row in board:
        if EMPTY in row:
            return False
    return True

def utility(board):
    win = winner(board)
    if win == X:
        return 10000
    elif win == O:
        return -10000

    score_x = 0
    score_o = 0

    for r in range(N):
        for c in range(N):
            if board[r][c] == X: score_x += 1
            elif board[r][c] == O: score_o += 1

    return score_x - score_o
````

**Bước 4: Định nghĩa hàm cắt tỉa Alpha-Beta**

Hàm `maxValue(state, alpha, beta, depth)` là phiên bản cải tiến của hàm maxValue trong Minimax thuần túy, được bổ sung thêm hai tham số alpha và beta để thực hiện cắt tỉa. Hàm này hoạt động tương tự như phiên bản gốc nhưng có thêm điều kiện kiểm tra: sau mỗi lần cập nhật giá trị v, hàm sẽ kiểm tra xem v có lớn hơn hoặc bằng beta không. Nếu có, điều này có nghĩa là MIN đã có một lựa chọn tốt hơn ở nhánh khác, nên không cần tiếp tục xét các nút con còn lại - đây chính là cơ chế cắt tỉa beta. Hàm cũng cập nhật alpha sau mỗi lần tìm được giá trị tốt hơn.

Hàm `minValue(state, alpha, beta, depth)` hoạt động tương tự nhưng theo hướng ngược lại. Sau khi cập nhật giá trị v, hàm kiểm tra xem v có nhỏ hơn hoặc bằng alpha không. Nếu có, điều này báo hiệu rằng MAX đã có một lựa chọn tốt hơn, nên có thể cắt tỉa nhánh này (cắt tỉa alpha). Hàm cũng cập nhật beta sau mỗi lần tìm được giá trị tốt hơn.

Hàm `minimax_alpha_beta(board, depth)` là hàm chính điều phối quá trình tìm kiếm với cắt tỉa. Hàm này nhận thêm tham số depth để giới hạn độ sâu tìm kiếm, điều này rất quan trọng với bàn cờ lớn vì không thể duyệt đến tận cùng. Hàm xác định người chơi hiện tại và gọi hàm tương ứng (maxValue hoặc minValue) với các giá trị alpha và beta khởi tạo là âm vô cùng và dương vô cùng. Nước đi nào cho giá trị tốt nhất sẽ được chọn làm nước đi cuối cùng.


````python
def maxValue(state, alpha, beta, depth):
    if depth == 0 or terminal(state):
        return utility(state)

    v = -math.inf

    for action in actions(state):
        v = max(v, minValue(result(state, action), alpha, beta, depth - 1))

        if v >= beta:
            return v

        alpha = max(alpha, v)

    return v

def minValue(state, alpha, beta, depth):
    if depth == 0 or terminal(state):
        return utility(state)

    v = math.inf

    for action in actions(state):
        v = min(v, maxValue(result(state, action), alpha, beta, depth - 1))

        if v <= alpha:
            return v

        beta = min(beta, v)

    return v

def minimax_alpha_beta(board, depth):
    current_player = player(board)
    best_move = None

    if current_player == X:
        v = -math.inf
        for action in actions(board):
            val = minValue(result(board, action), -math.inf, math.inf, depth - 1)
            if val > v:
                v = val
                best_move = action

    else:
        v = math.inf
        for action in actions(board):
            val = maxValue(result(board, action), -math.inf, math.inf, depth - 1)
            if val < v:
                v = val
                best_move = action

    return best_move
````

**Bước 5: Chương trình chính**

Chương trình chính của phần Alpha-Beta được thiết kế tương tự như phần Minimax nhưng có thêm các tính năng linh hoạt hơn. Đầu tiên, chương trình yêu cầu người dùng nhập cả kích thước bàn cờ N và số quân liên tiếp cần để thắng (WIN_COUNT). Điều này cho phép người dùng tùy chỉnh độ khó và thời gian chơi của trò chơi. Sau khi khởi tạo, chương trình thiết lập AI là X và cho AI đi trước. Vòng lặp chính sẽ tiếp tục cho đến khi game kết thúc, với mỗi lượt AI sẽ sử dụng thuật toán Alpha-Beta với độ sâu giới hạn là 2 để tính toán nước đi. Việc giới hạn độ sâu là cần thiết vì với bàn cờ lớn và việc sử dụng deepcopy, việc duyệt sâu hơn sẽ tốn rất nhiều thời gian và bộ nhớ. Khi trò chơi kết thúc, chương trình sẽ hiển thị kết quả dựa trên giá trị utility, với các ngưỡng 1000 và -1000 để phân biệt giữa thắng/thua và hòa.


````python
if __name__ == "__main__":
    # 0. Nhập kích thước bàn cờ và số quân cần để thắng
    global N, WIN_COUNT
    while True:
        try:
            N = int(input("Nhập kích thước bàn cờ N (ví dụ: 5 cho 5x5): "))
            if N >= 3:
                break
            else:
                print("Kích thước bàn cờ phải lớn hơn hoặc bằng 3. Vui lòng nhập lại.")
        except ValueError:
            print("Vui lòng nhập một số nguyên hợp lệ.")

    while True:
        try:
            WIN_COUNT = int(input(f"Nhập số quân liên tiếp cần để thắng (tối đa {N}): "))
            if 3 <= WIN_COUNT <= N:
                break
            else:
                print(f"Số quân cần để thắng phải từ 3 đến {N}. Vui lòng nhập lại.")
        except ValueError:
            print("Vui lòng nhập một số nguyên hợp lệ.")

    # Khởi tạo bàn cờ
    board = initial_state()
    print(f"\n--- TIC-TAC-TOE {N}x{N} (ALPHA-BETA, DEEPCOPY) ---")
    print(f"Luật thắng: {WIN_COUNT} quân liên tiếp.")

    # Thiết lập: AI là X (đi trước)
    ai_turn = True
    print("AI đóng vai X. AI đi trước.")

    while True:
        # 1. In bàn cờ hiện tại
        print("\nBàn cờ hiện tại:")
        print(np.array(board))

        # 2. Kiểm tra kết thúc game
        if terminal(board):
            score = utility(board)
            if score >= 1000:
                print("\n>> KẾT QUẢ: AI (X) THẮNG!")
            elif score <= -1000:
                print("\n>> KẾT QUẢ: BẠN (O) THẮNG!")
            else:
                print("\n>> KẾT QUẢ: HÒA!")
            break

        # 3. Xử lý lượt đi
        if ai_turn:
            print("\nAI đang suy nghĩ (Độ sâu 2)...")
            # Lưu ý: Do dùng deepcopy nên rất chậm, chỉ nên để depth=2
            move = minimax_alpha_beta(board, depth=2)

            if move:
                board = result(board, move)
                print(f"AI đã đánh vào ô: {move}")
            else:
                print("AI không tìm thấy nước đi (Lỗi hoặc Hòa).")
                break
            ai_turn = False # Chuyển lượt cho người chơi

        else:
            # Lượt người chơi
            try:
                print("\nLượt của BẠN (O). Hãy nhập tọa độ:")
                r = int(input(f"   Nhập hàng (0-{N-1}): "))
                c = int(input(f"   Nhập cột (0-{N-1}): "))

                # Kiểm tra nước đi hợp lệ
                if (r, c) in actions(board):
                    board = result(board, (r, c))
                    ai_turn = True # Chuyển lượt cho AI
                else:
                    print("Lỗi: Nước đi không hợp lệ hoặc ô đã có quân.")
            except ValueError:
                print("Lỗi: Vui lòng nhập số nguyên.")
````
